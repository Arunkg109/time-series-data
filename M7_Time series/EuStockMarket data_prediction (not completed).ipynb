```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# test stationary
from matplotlib import pyplot as plt

plt.plot(df)
```

```python
rolmean = df.rolling(window = 15).mean()

# Lets plot the rolling average 

plt.plot(rolmean, label = "Moving Average")
plt.show()
```

```python
# data is not stationary in nature
```

```python
# Converting the non-stationary time series into a stationary time series
```

```python
# peform the log tranformation

log_ts = np.log(df)

log_ts
```

```python
# differencing the time series - This can be done using shift()

diff1_ts = log_ts - log_ts.shift()
```

```python
# plot the diff1_ts

plt.plot(diff1_ts)
```

```python
#  look like that its a stationary time series
```

```python
# Lets build the ARIMA model

from statsmodels.tsa.arima_model import ARIMA

# Model

model_ma = ARIMA(log_ts, order = (0,1,1))
```
